```python
# import require libararies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',None)
```

```python
# Read CSV file
df = pd.read_csv('/content/HR_Employee_Attrition.csv')
# Show first five rows
df.head()
```

```python
# Provide information about number of rows and columns
df.shape
```

```python
#Gives Data-Types and Non -null values from columns
df.info()
```

```python
# check for null values
df.isnull().sum()
```

```python
# Provide satistical summary
df.describe()
```

```python
# Provide name of all columns from dataset
df.columns
```

```python
# Shows the actual value count
df['Attrition'].value_counts()
```

```python
# Shows Percentage Distribution and gives information about data is imbalanced or not
df['Attrition'].value_counts(normalize=True)*100
```

```python
sns.countplot(x='Attrition',data=df)
plt.title('Employee Attrition Count')
plt.show()
```

```python
attrition_rate=df['Attrition'].value_counts(normalize=True)*100
attrition_rate
```

```python
# Now we plot the graph for Attriton By Department
# R&D & Sales department has highest attrition
plt.figure()
sns.countplot(x='Department',hue='Attrition',data=df)
plt.title('Attrition by Department')
plt.show()
```

```python
#Attrition by gender
# Male candidates have higher attrition rate
plt.figure()
sns.countplot(x='Gender',hue='Attrition',data=df)
plt.title('Attrition By Gender')
plt.show()
```

```python
# Now we find can overtime work actual affect on the attrition rate
# Yes overtime affect attrition significantly
plt.figure()
sns.countplot(x='OverTime',hue='Attrition',data=df)
plt.title('Attrition Vs Overtime')
plt.show()
```

```python
#Atttrition by job role
# Laboratory Sales Executive & Technician has highest Attrition rate
plt.figure()
sns.countplot(y='JobRole',hue='Attrition',data=df)
plt.title('Attrition by job role')
plt.show()
```

```python
# Attrition rate Vs Monthly Income
# Employee with lower salary tends to leave the company
plt.figure()
sns.barplot(x='Attrition',y='MonthlyIncome',data=df)
plt.title('Attrition Vs Monthly Income')
plt.show()
```

```python
# Attrition Vs Job Satisfaction
#Low Satisfaction tends to Higher Attrition
plt.figure()
sns.countplot(x='JobSatisfaction',hue='Attrition',data=df)
plt.title('Attrition Vs Job Satisfaction')
plt.show()
```

```python
# Attrition Vs Work-Life Balance
# Poor work-life balance strongly contributes to attrition.
plt.figure()
sns.countplot(x='WorkLifeBalance',hue='Attrition',data=df)
plt.title('Work-Life Balance Vs Attrition')
plt.show()
```

```python
# Attrition Vs Years At Company
# Employee leaves mostly in early years
plt.figure()
sns.barplot(x='Attrition',y='YearsAtCompany',data=df)
plt.title('Attrition Vs Years At Company')
plt.show()
```

```python
# Correlation Heatmap(Numerical Features)

plt.figure(figsize=(12,8))
corr = df.select_dtypes(include=['int64','float64']).corr()
sns.heatmap(corr,cmap='coolwarm',linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
